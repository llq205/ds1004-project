```python
#ADDR_PCT_CD

labels = ['VALID', 'NULL']
sizes = [5100841, 390]

colors = ['lightblue','yellow']
explode = (0.1, 0)  # explode 1st slice
 
    
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct=make_autopct(sizes), shadow=True,startangle=140)
 
plt.axis('equal')
plt.title('ADDR_PCT_CD')
# plt.savefig('ADDR_PCT_CD.png')
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("NYC_Income.csv", index_col = 0)
data = data.T
data = data.ix[:,0:5]
data = data.reset_index()
data.columns.values[0] = 'Year' 
data = pd.DataFrame(data)

pos = list(range(len(data['Bronx'])))
width = 0.18

# Plotting the bars
plt.figure()
fig, ax = plt.subplots(figsize=(20,10))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos,
        #using df['pre_score'] data,
        pd.to_numeric(data['Bronx']),
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#EE3224',
        # with label the first value in first_name
        label=str(data['Year'][0]))

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos],
        #using df['mid_score'] data,
        pd.to_numeric(data['Brooklyn (Kings)']),
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#F78F1E',
        # with label the second value in first_name
        label=data['Year'][1])

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos],
        #using df['post_score'] data,
        pd.to_numeric(data['Manhattan']),
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#FFC222',
        # with label the third value in first_name
        label=data['Year'][2])


plt.bar([p + width*3 for p in pos],
        pd.to_numeric(data['Queens']),
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#ee247b',
        # with label the first value in first_name
        label=data['Year'][3])


plt.bar([p + width*4 for p in pos],
        #using df['mid_score'] data,
        pd.to_numeric(data['Staten Island (Richmond)']),
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#dd24ee',
        # with label the second value in first_name
        label=data['Year'][4])


# Set the y axis label
ax.set_ylabel('Income')

# Set the chart's title
ax.set_title('Personal Income in NYC')

# Set the position of the x ticks
ax.set_xticks([p + 2.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(data['Year'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*6)
plt.ylim([0, max(pd.to_numeric(data['Bronx']) + pd.to_numeric(data['Brooklyn (Kings)']) + pd.to_numeric(data['Manhattan']) + pd.to_numeric(data['Queens']) + pd.to_numeric(data['Staten Island (Richmond)']))] )

# Adding the legend and showing the plot
plt.legend(['Bronx','Brooklyn (Kings)','Manhattan','Queens','Staten Island (Richmond)'], loc='upper left')
plt.grid()
plt.savefig('C:/Users/Aileen/Desktop/results/NY_Income.png')
plt.show()
```
